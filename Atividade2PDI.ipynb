````python
# Questão 3

## Caderno Colab: https://colab.research.google.com/drive/1_TSYvAwj6belPgx-rNQ7W0ggM2HaDIxp?authuser=1#scrollTo=OGuyICFQwWNx

# Importando bibliotecas
import cv2 as cv
import numpy as np

# Funções
## Lendo a matriz inicial ou o kernel - Código tirado do Moodle da disciplina (exemplo2 - erosão 2D)
def lerImagem():
  f = []
  ler_linha = input()
  while ler_linha: 
    f.append([int(i) for i in ler_linha.split(' ') if i])
    ler_linha = input()
  return np.array(f).astype('uint8')

## Imprimindo a matriz com a abertura -  Código tirado do Moodle da disciplina (exemplo2 - erosão 2D)
def imprimirImagem(m):
  [l,c] = m.shape
  st = ''
  for i in range(l):
    for j in range(c):
      st += str(m[i][j]) + ' '
    st += '\n'
  return st

# Código
## Lendo kernel
kernel = lerImagem()

## Lendo imagem
imagem = lerImagem()

## Aplicando primeiro a erosão (erode) e depois a dilatação (dilate) da erosão para compor a abertura
erosao = cv.erode(imagem, kernel)
abertura = cv.dilate(erosao, kernel) 

#Imprimindo a imagem após tratamento
print(imprimirImagem(abertura))
````

#### Questão 4

Considere a dilatação definida por:
Captura de ecrã de 2022-05-09 18-59-31.png

Considere também a erosão definida por:
Captura de ecrã de 2022-05-09 19-00-01.png

Resolver utilizando composições de dilatação e erosão do cv2: ```cv2.erode``` e ```cv2.dilate```.

````python
# Questão 4

## Caderno Colab: https://colab.research.google.com/drive/1_TSYvAwj6belPgx-rNQ7W0ggM2HaDIxp?authuser=1#scrollTo=OGuyICFQwWNx

# Importando bibliotecas
import cv2 as cv
import numpy as np

# Funções
## Lendo a matriz inicial ou o kernel - Código tirado do Moodle da disciplina (exemplo2 - erosão 2D)
def lerImagem():
  f = []
  ler_linha = input()
  while ler_linha: 
    f.append([int(i) for i in ler_linha.split(' ') if i])
    ler_linha = input()
  return np.array(f).astype('uint8')

## Imprimindo a matriz com o gradiente -  Código tirado do Moodle da disciplina (exemplo2 - erosão 2D)
def imprimirImagem(m):
  [l,c] = m.shape
  st = ''
  for i in range(l):
    for j in range(c):
      st += str(m[i][j]) + ' '
    st += '\n'
  return st

# Código
## Lendo kernel
kernel = lerImagem()

## Lendo imagem
imagem = lerImagem()

## Aplicando primeiro a dilatação (dilate) e depois a erosão (erode) na imagem inicial para compor o gradiente (dilatacao - erosão)
dilatacao = cv.dilate(imagem, kernel) 
erosao = cv.erode(imagem, kernel)
gradiente = dilatacao - erosao

#Imprimindo a imagem após tratamento
print(imprimirImagem(gradiente))
````
